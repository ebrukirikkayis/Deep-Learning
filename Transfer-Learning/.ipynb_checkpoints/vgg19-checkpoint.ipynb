```python
from keras.applications.vgg19 import VGG19
from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Dense, Flatten
from keras.datasets import cifar10
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
(x_train, y_train),(x_test, y_test)= cifar10.load_data()
print("x_train shape", x_train.shape)
print("train sample", x_train.shape[0])
```

```python
num_of_class=10
```

```python
y_train= to_categorical(y_train,num_of_class)
y_test= to_categorical(y_test, num_of_class)
```

```python
input_shape= x_train.shape[1:]
input_shape
```

```python
# visualize
plt.imshow(x_train[510].astype(np.uint8))
plt.show()
```

```python
# vgg19 inputları 48x48 olmalı bizim inputlarımız 32x32 bunu reshape etmemiz gerekiyor
# increase dimension
def resize_img(img):
    num_of_img= img.shape[0]  # 5000 tane görseli al
    new_array= np.zeros((num_of_img, 48,48,3))  # her biri için bir alan yarat
    for i in range(num_of_img):
        new_array[i]= cv2.resize(img[i,:,:,:],(48,48))
    return new_array
```

```python
x_train = resize_img(x_train)
x_test= resize_img(x_test)
```

```python
print("increased dim x_train", x_train.shape)
```

```python
plt.figure()
plt.imshow(x_train[510].astype(np.uint8))
plt.show()
```

```python
vgg= VGG19(include_top= False, weights= "imagenet", input_shape=(48,48,3))
#VGG19'un imagenet ile eğitilmiş ağırlıklarını kullan ve include_top false diyerek fc layer'ı eklememiş oluyoruz.
print(vgg.summary())
```

```python
vgg_layer_list= vgg.layers
print(vgg_layer_list)
```

```python
model=Sequential()
```

```python
for layer in vgg_layer_list:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable= False

# modelde bulunan ve önceden eğitilmiş layer'ları train etmiyoruz.
```

```python
# fully con. layers
model.add(Flatten())
model.add(Dense(128))
model.add(Dense(num_of_class, activation="softmax"))

# bu katmanları ekledikten sonra train yapılacak
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",
             optimizer="rmsprop",
             metrics= ["accuracy"])
```

```python
hist= model.fit(x_train, y_train, validation_split= 0.2,
               epochs= 2, batch_size=512)
```

```python
model.save_weights("vgg19-cifar10.h5")
```

```python
plt.plot(hist.history["loss"], label="train loss")
plt.plot(hist.history["val_loss"], label="val loss")
plt.legend()
plt.show()
```

```python
plt.plot(hist.history["accuracy"], label="train accuracy")
plt.plot(hist.history["val_accuracy"], label="val accuracy")
plt.legend()
plt.show()
```

```python
import json, codecs
with open("vgg19-cifar10-hist.json","w", encoding="utf-8") as f:
    json.dump(hist.history,f)
```

```python
#load history
with codecs.open("vgg19-cifar10-hist.json","r", encoding="utf-8") as f:
    n= json.loads(f.read())
    
plt.plot(n["loss"], label="training loss")
plt.plot(n["val_loss"], label="validation loss")
plt.legend()
plt.show()


plt.plot(n["accuracy"], label="training accuraccy")
plt.plot(n["val_accuracy"], label="validation accuracy")
plt.legend()
plt.show()
```
